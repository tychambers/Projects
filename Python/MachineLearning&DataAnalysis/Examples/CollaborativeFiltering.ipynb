```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings('ignore')
```

```python
bank_full = pd.read_csv('bank_full.csv')
bank_full.head()
```

```python
bank_full.info()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

# convert to a number so we can feed to machine learning algorithm
bank_full = bank_full.apply(le.fit_transform)
bank_full.head()
```

```python
bank_full
```

```python
@ the row names are really long
new_bank_full = bank_full.loc[:,'housing_loan                ':'divorced                    ']
```

```python
new_bank_full
```

```python
X = new_bank_full
# below 1 means will qualify for the loan, 0 means will not
y = bank_full['y']
```

```python
LogReg = LogisticRegression()
LogReg.fit(X, y)
```
