```python
# Below reads a csv to a dataset, then puts the contents of that dataset into a table in an sqlite db file
```

```python
import pandas as pd

df = pd.read_csv('imdb_movies.csv')
df
```

```python
import sqlite3 as sq
data = df
sql_data = 'C:\\users\\tyler\\movies.db' #- Creates DB names SQLite
conn = sq.connect(sql_data)
cur = conn.cursor()
cur.execute('''DROP TABLE IF EXISTS Movies''')
data.to_sql('Movies', conn, if_exists='replace', index=False) # - writes the pd.df to SQLIte DB
pd.read_sql('select * from Movies', conn)
conn.commit()
conn.close()
```
